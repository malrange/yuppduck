```python
import pandas as pd
import re
import time
from datetime import datetime, date, timedelta
from bs4 import BeautifulSoup
import requests

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 네이버 리뷰 크롤링 함수

def naver_review_crawl(code):
    global id
    global content
    global date
    global name
    url = f'https://pcmap.place.naver.com/restaurant/{code}/review/visitor'
    browser = webdriver.Edge() #크롬쓰시면 수정하세요!
    browser.get(url)
    browser.implicitly_wait(5)
    time.sleep(3)
    while True:
        try :
            browser.find_element(By.CSS_SELECTOR, '#app-root > div > div > div > div:nth-child(6) > div:nth-child(3) > div.place_section.k5tcc > div.NSTUp > div > a').click() #더보기
            time.sleep(2)
        except :
            break


    html = browser.page_source
    data = BeautifulSoup(html, 'lxml')
    name2 = data.select_one('span.Fc1rA').text
    id_lst = data.select('div.VYGLG')
    content_lst = data.select('span.zPfVt')
    date_lst = data.select('div.qM6I7 > div > div._7kR3e > span:nth-child(1) > span:nth-child(3)')

    for i, c, d in zip(id_lst, content_lst, date_lst) :
        name.append(name2)
        id.append(i.text)
        content.append(c.text.replace('\n', ' ').strip())
        date.append(d.text)
```

```python
# 강남구 매장 리뷰 크롤링

gangnam = [1065745589, 36601251, 36797355, 37536954, 33586177, 37003739, 36480598, 1839198466]
name, id, content, date = [], [], [], []
#print(i, '크롤링 시작')
for i in gangnam:
    naver_review_crawl(i)
df = pd.DataFrame({
    'LOCATION' : name,
    'ID' : id,
    'Review' : content,
    'Date' : date
})
filename = f"{datetime.now().strftime('%Y-%m-%d')} '강남구'.xlsx"
df.to_excel(filename, index=False)
```
